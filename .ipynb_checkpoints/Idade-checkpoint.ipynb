```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
from scipy.stats import ks_2samp

import warnings
warnings.filterwarnings('ignore')

import matplotlib as mpl
from matplotlib import pyplot as plt
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
from matplotlib.ticker import FormatStrFormatter
import seaborn as sns
palette = 'viridis'
font = {'family': 'serif', 'color':  'black', 'weight': 'normal','size': 14}
font_cb = {'family': 'serif', 'color':  'black', 'weight': 'normal','size': 12}
```

```python

data = pd.read_csv('data.csv', index_col='Cluster')
df = pd.read_csv('log-results-eDR3-MF_integrada.csv', index_col='Cluster')
```

```python
def classificar_idade(age):
    if age < 8:
        return 'Young clusters'
    #elif 7.5 <= age < 8.5:
    #    return 'Intermediate-age clusters'
    else:
        return 'Old-clusters'

# Aplicar a função e criar/atualizar coluna
data['age_group'] = data['age'].apply(classificar_idade)
df['age_group'] = df['age'].apply(classificar_idade)
aux=data[data.q>0]
```

```python
fig, axs = plt.subplots(1, 2 , figsize=(10,4), sharey='row')
i=0
age_order = ['Young clusters', 'Old-clusters']
for group in age_order:
    sns.histplot(data = aux[aux.age_group == group], x='q',  element='step', bins=12, stat= 'percent', ax=axs[i], legend=False)
    axs[i].set_title(group, fontdict=font)
    axs[i].set_xlabel('Mass ratio', fontdict=font)
    i+=1
    print(len(aux[aux.age_group == group]))
axs[0].set_ylabel('Percent [%]', fontdict=font)
    
#plt.savefig('hist_mass_ratio_mass.png',bbox_inches='tight', dpi=300)
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.scatterplot(data=df, x='age', y= 'bin_frac', hue='mean_stellar_mass', palette=palette, legend=False)
sns.regplot(data=df, x='age', y='bin_frac',
            scatter=False, )



plt.errorbar(df.age, df.bin_frac, yerr=[df.e_fb_i, df.e_fb_s], 
             fmt='none', ecolor='lightgray', elinewidth=.5, capsize=.5, alpha=.3
)



plt.xlabel('log(age) [yr]', fontdict=font)
plt.ylabel('Binary fraction', fontdict=font)

norm = mpl.colors.Normalize(vmin=df['mean_stellar_mass'].min(), vmax=df['mean_stellar_mass'].max()) 
sm = plt.cm.ScalarMappable(cmap=palette, norm=norm) 
sm.set_array([]) 

cbar = plt.colorbar(sm, ticks=np.arange(0, 5,1),) 
cbar.set_label('Título do Colorbar', labelpad=2)

cbar.ax.yaxis.set_label_position('left') 
cbar.ax.set_ylabel(r'Median stellar mass [$M_\odot$]', rotation=90, labelpad=-55, va='bottom', fontdict= font_cb )


plt.savefig('dist_fb_age.png',bbox_inches='tight', dpi=300)
plt.show()

```

```python
data['mass_system'] = data.mass+data.comp_mass
```

```python
df['mean_stellar_mass'] = np.nan
for cluster in df.index:
    df.loc[cluster, 'mean_stellar_mass'] = data.loc[cluster, 'mass_system'].sum()/(len(data.loc[cluster])+len(data[data.q>0].loc[cluster]))
```

```python
from scipy.stats import ks_2samp

# Suponha que x1 e x2 são suas duas amostras
stat, p_value = ks_2samp(aux[aux.age_group =='Young clusters'].q, aux[aux.age_group =='Old-clusters'].q)

print(f'Estatística KS = {stat:.3f}, p-valor = {p_value:.3f}')
if p_value < 0.05:
    print("As distribuições são significativamente diferentes (nível 5%)")
else:
    print("Não há evidência suficiente para dizer que são diferentes")
```
