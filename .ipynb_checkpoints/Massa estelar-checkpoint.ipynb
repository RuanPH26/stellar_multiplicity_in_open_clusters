```python
# Criar faixas de massa
mass_bins = range(int(data['mass'].min()), int(data['mass'].max()) + 2, 1)
data['mass_bin'] = pd.cut(data['mass'], bins=mass_bins, right=False)

# Filtrar bins com >100 estrelas
counts_per_bin = data['mass_bin'].value_counts().sort_index()
valid_bins = counts_per_bin[counts_per_bin > 50].index
filtered_data = data[data['mass_bin'].isin(valid_bins)]

# Contar sistemas por faixa de massa e tipo
grouped = (
    filtered_data
    .groupby(['mass_bin', 'system_label'])
    .size()
    .reset_index(name='count')
)

# Calcular total por faixa de massa
grouped['total_per_bin'] = grouped.groupby('mass_bin')['count'].transform('sum')

# Calcular porcentagem
grouped['percentage'] = (grouped['count'] / grouped['total_per_bin'])

# Se mass_bin for categórico, removemos os níveis não usados
if pd.api.types.is_categorical_dtype(grouped['mass_bin']):
    grouped['mass_bin'] = grouped['mass_bin'].cat.remove_unused_categories()

# Filtrar apenas bins com dados
grouped = grouped[grouped['total_per_bin'] > 0].copy()

# Criar rótulos legíveis para os bins
def format_bin_label(interval):
    left = int(interval.left)
    right = int(interval.right)
    return f"{left}–{right}"

# Se ainda for Interval, aplicar formatação
if isinstance(grouped['mass_bin'].iloc[0], pd._libs.interval.Interval):
    grouped['mass_bin_label'] = grouped['mass_bin'].apply(format_bin_label)
else:
    grouped['mass_bin_label'] = grouped['mass_bin'].astype(str)
if pd.api.types.is_categorical_dtype(grouped['mass_bin_label']):
    grouped['mass_bin_label'] = grouped['mass_bin_label'].cat.remove_unused_categories()

# Agora plote com os rótulos formatados
plt.figure(figsize=(12, 6))
sns.barplot(
    data=grouped[grouped['system_label'] == 'binaries'],
    x='mass_bin_label',
    y='percentage',
    palette=palette,

)


plt.xlabel(r'$M_1$' + ' '+ r'$[M_\odot]$', fontdict=font)
plt.ylabel(r'$f_b^{ prim}$', fontdict=font)
plt.legend(title='')
plt.tight_layout()
plt.show()
```

# Calcular a presença de estrelas de uma dada faixa de massa em sistemas binários, seja como estrela principal ou como secundária

1° Selecionar todos os sistemas que contemplem estrelas de uma dada faixa de massa, seja ela como estrela principal ou secundária

2° Determinar a quantidade desses sistemas

3° Determinar nesses sistemas quantos deles a estrela principal é dessa dada faixa de massa e determinar a fb nesses casos

4° Determinar quantos sistemas possuem uma estrela de uma faixa de massa como secundária, e estimar a frequência dessas 
estrelas em sistemas binários através da relação: $M_f = \frac{primarias + secundarias}{\text{Total de estrelas}}$

em que total de sistemas é a soma das estrelas de dada massa como primárias, como secundárias e isoladas

# Low mass stars: $M<0.7M_\odot$

```python
low_mass = data[
    (data.mass < 0.7) |
    ((data.comp_mass > 0) & (data.comp_mass < 0.7))
]

#Quantidade de primárias de baixa massa
bin_prim_lm = low_mass[(low_mass.mass<0.7) & low_mass.comp_mass>0].shape[0]

#Quantidade de secundárias de baixa massa
bin_sec_lm = low_mass[(low_mass.comp_mass>0) & (low_mass.comp_mass<0.7)].shape[0]

#Total de estrelas de baixa massa
total_lm = low_mass[low_mass.mass<0.7].shape[0] + low_mass[(low_mass.comp_mass>0) & (low_mass.comp_mass<0.7)].shape[0]

#fração de binárias considerando apenas sistemas em que a primária é de baixa massa
fb_prim_lm = frac_bin(low_mass[low_mass.mass<0.7])

mf_lm = (bin_prim_lm+bin_sec_lm)/total_lm

companion_frequence_lm = bin_sec_lm/total_lm
```

# Solar type: $0.7\leq M<1.5M_\odot$


```python
solar_type = data[
    ((data.mass >= 0.7) & (data.mass<1.5)) |
    ((data.comp_mass >= 0.7) & (data.comp_mass<1.5)
)
]

#Quantidade de primárias de baixa massa
bin_prim_st = solar_type[(solar_type.mass<1.5) & solar_type.comp_mass>0].shape[0]

#Quantidade de secundárias de baixa massa
bin_sec_st = solar_type[(solar_type.comp_mass>=0.7) & (solar_type.comp_mass<1.5)].shape[0]

#Total de estrelas de baixa massa
total_st = solar_type[solar_type.mass<1.5].shape[0] + solar_type[(solar_type.comp_mass>=0.7) &(solar_type.comp_mass<1.5)].shape[0]

#fração de binárias considerando apenas sistemas em que a primária é de baixa massa
fb_prim_st = frac_bin(solar_type[(solar_type.mass >= 0.7) & (solar_type.mass<1.5)])

mf_st = (bin_prim_st+bin_sec_st)/total_st

companion_frequence_st = bin_sec_st/total_st
```

# Intermediate mass: $1.5\leq M< 8.0 M_\odot$

```python
intermediate_mass = data[
    ((data.mass >= 1.5) & (data.mass<8)) |
    ((data.comp_mass >= 1.5) & (data.comp_mass<8)
)
]

#Quantidade de primárias de baixa massa
bin_prim_im = intermediate_mass[(intermediate_mass.mass<8) & intermediate_mass.comp_mass>0].shape[0]

#Quantidade de secundárias de baixa massa
bin_sec_im = intermediate_mass[(intermediate_mass.comp_mass>=1.5) & (intermediate_mass.comp_mass<8)].shape[0]

#Total de estrelas de baixa massa
total_im = intermediate_mass[intermediate_mass.mass<8].shape[0] + intermediate_mass[(intermediate_mass.comp_mass>=1.5) &(intermediate_mass.comp_mass<8)].shape[0]

#fração de binárias considerando apenas sistemas em que a primária é de baixa massa
fb_prim_im = frac_bin(intermediate_mass[(intermediate_mass.mass >= 1.5) & (intermediate_mass.mass<8)])

mf_im = (bin_prim_im+bin_sec_im)/total_im

companion_frequence_im = bin_sec_im/total_im
```

# High mass: M $\geq 8M\odot$

```python
high_mass = data[
    (data.mass >=8) |
    ((data.comp_mass > 0) & (data.comp_mass >=8))
]

#Quantidade de primárias de baixa massa
bin_prim_hm = high_mass[(high_mass.mass>=8) & high_mass.comp_mass>0].shape[0]

#Quantidade de secundárias de baixa massa
bin_sec_hm = high_mass[(high_mass.comp_mass>0) & (high_mass.comp_mass>=8)].shape[0]

#Total de estrelas de baixa massa
total_hm = high_mass[high_mass.mass>8].shape[0] + high_mass[(high_mass.comp_mass>0) & (high_mass.comp_mass>=8)].shape[0]

#fração de binárias considerando apenas sistemas em que a primária é de baixa massa
fb_prim_hm = frac_bin(high_mass[high_mass.mass>=8])

mf_hm = (bin_prim_hm+bin_sec_hm)/total_hm

companion_frequence_hm = bin_sec_hm/total_hm
```

```python
m = 0
mass_prim = unp.uarray([],[])
mass_range = unp.uarray([],[])
bin_prim = unp.uarray([],[])
bin_sec = unp.uarray([],[])
mf = unp.uarray([],[])
q = [[], [], [], []]
dm=1

lista_fb = [bin_prim, bin_sec, mf, ]
row=0

q_std = [0.45, 0.65, 0.85, 1]
for q_val in q_std:
    
    data_ = data[data.q<q_val]


    while m < data_.mass.max():
        aux = data_[((data_.mass > m) & (data_.mass<=m+dm)) | ((data_.comp_mass > m) & (data_.comp_mass<= m+dm))]
        if len(aux)>=100:

            prim_mass = unp.uarray(aux.mass, aux.er_mass).mean()        
            nominais = np.append(unp.nominal_values(mass_prim), unp.nominal_values(prim_mass))
            desvios = np.append(unp.std_devs(mass_prim), unp.std_devs(prim_mass))
            mass_prim = unp.uarray(nominais, desvios)

            comp_mass = unp.uarray(aux[(aux.comp_mass>m) & (aux.comp_mass<m+1)].comp_mass, 
                                      aux[(aux.comp_mass>m) & (aux.comp_mass<m+1)].er_comp_mass).mean()

            nominais = np.append(unp.nominal_values(comp_mass), unp.nominal_values(prim_mass))
            desvios = np.append(unp.std_devs(comp_mass), unp.std_devs(prim_mass))

            mean_mass = unp.uarray(nominais, desvios).mean()
            nominais = np.append(unp.nominal_values(mass_range), unp.nominal_values(mean_mass))
            desvios = np.append(unp.std_devs(mass_range), unp.std_devs(mean_mass))
            mass_range = unp.uarray(nominais, desvios)

            n_prim  = aux[(aux.mass<=m+1) & aux.comp_mass>0].shape[0] #Quantidade de primárias
            n_sec = aux[(aux.comp_mass>m) & (aux.comp_mass<=m+dm)].shape[0] #Quantidade de secundárias
            total = aux[aux.mass<=m+1].shape[0] + aux[(aux.comp_mass>m) & (aux.comp_mass<=m+1)].shape[0]  #Total de estrelas

            # Erros associados a contagem para cada caso

            sigma_sec = np.sqrt(n_sec)
            sigma_total = np.sqrt(total)

            n_prim =ufloat(n_prim, np.sqrt(n_prim))
            n_sec = ufloat(n_sec, sigma_sec)
            total = ufloat(total, sigma_total)


            bin_mask = (aux.mass > m) & (aux.mass <= m + dm)
            n_bin = ufloat(len(aux[bin_mask & (aux.q > 0)]), np.sqrt(len(aux[bin_mask & (aux.q > 0)])))
            n_sing = ufloat(len(aux[bin_mask & (aux.q == 0)]), np.sqrt(len(aux[bin_mask & (aux.q == 0)])))


            fb_prim = n_bin / (n_bin+n_sing)





            fb_sec = n_sec/total
            fb_mf = (n_prim+n_sec)/total

            var = [fb_prim,fb_sec, fb_mf, bin_frac_m]

            for i in range(3):
                lista_fb[i] = unp.uarray(
                    np.append(unp.nominal_values(lista_fb[i]), unp.nominal_values(var[i])),
                    np.append(unp.std_devs(lista_fb[i]), unp.std_devs(var[i]))
                )

            q[0].append(aux[(aux.mass > m) & (aux.mass<=m+dm) & (aux.q>0)].q.median())
            q[1].append(aux[aux.q>0].q.median())
            q[2].append(aux[(aux.comp_mass > m) & (aux.comp_mass<=m+dm)].q.median())
            q[3].append(aux[(aux.mass > m) & (aux.mass<=m+dm)].q.median())

        m+=.5
    bin_prim, bin_sec, mf = lista_fb
    print(q_val)
    fig, axs = plt.subplots(4, 3, figsize=(16, 16), )
    plot()
    row+=1
    
plt.show()
```

```python
def calcula_sistemas(df, m, dm=1):
    ns = len(df[df.q==0])
    n_prim = len(df[(df.mass>m) & (df.mass<= m +dm) & (df.q>0)])
    n_sec = len(df[(df.comp_mass>m) & (df.comp_mass<= m +dm)])
    total = len(df[((df.mass>m) & (df.mass<= m +dm) )| ((df.comp_mass>m) & (df.comp_mass<= m +dm))])
    return ns, n_prim, n_sec, total

def calcula_fb_prim(df, m, dm=1):
    total = len(df)
    N_prim = len(df[(df.mass>= m) & (df.mass<m+dm) & (df.comp_mass>0)])
    return N_prim/total

def calcula_fb_sec(df, m, dm=1):
    total = len(df)
    n_sec = len(df[(df.comp_mass>= m) & (df.comp_mass<m+dm) & (df.q>0)])
    return n_sec/total

def calcula_mf(df,m, dm=1):
    n_prim = len(df[(df.mass>m) & (df.mass<= m +dm) & (df.q>0)])
    n_sec = len(df[(df.comp_mass>m) & (df.comp_mass<= m +dm)])
    total =len(df)
    return (n_prim+n_sec)/total






def calcula_probabilidades(df, stat,dm=1):
    m =0
    x = []
    er_x = []
    fb_prim = []
    fb_sec = []
    mf = []
    n_stars =[]
    
    while m < df.mass.max():
        
        mask = ((df.mass >= m) & (df.mass < m + dm)) | \
               ((df.comp_mass >= m) & (df.comp_mass < m + dm) & (df.q > 0))

        aux = df[mask]

        if len(aux) >= 100:
            primary_mass = aux[(aux.mass >= m) & (aux.mass < m + dm)][['mass', 'er_mass']]
            comp_mass = aux[(aux.comp_mass >= m) & (aux.comp_mass < m + dm)][['comp_mass', 'er_comp_mass']]
            
            pmemb = np.concatenate((aux[(aux.mass >= m) & (aux.mass < m + dm)]['Pmemb'],
                                   aux[(aux.comp_mass >= m) & (aux.comp_mass < m + dm)]['Pmemb']))

            all_masses = np.concatenate((primary_mass.mass.values, comp_mass.comp_mass.values))
            er_all_masses = np.concatenate((primary_mass.er_mass.values, comp_mass.er_comp_mass.values))
            
            all_masses = unp.uarray(all_masses, er_all_masses)
            
            n_stars.append(len(all_masses))
            
            
            
            
            if stat=='media':
                
                mean = all_masses.mean()
                x.append(unp.nominal_values(mean))
                er_x.append(unp.std_devs(mean))
                
            elif stat== 'ponderada':
                
             
                
                pesos = pmemb*(1/(unp.std_devs(all_masses)+1e-10)**2)
         
                
                
                numerador = (all_masses*pesos).sum()
          
                denominador = pesos.sum()
      
                mean = numerador/denominador
        
                x.append(unp.nominal_values(mean))
                er_x.append(unp.std_devs(mean))              
                

            fb_prim.append(calcula_fb_prim(aux,m, dm))
            fb_sec.append(calcula_fb_sec(aux,m, dm))
            mf.append(calcula_mf(aux,m, dm))

        m += dm
    return np.array(x), np.array(er_x), np.array(fb_prim), np.array(fb_sec), np.array(mf), n_stars
```

```python
x, er_x, fb_prim, fb_sec, mf, n_stars_all = calcula_probabilidades(data, 'media',1)
#x, er_x, fb_prim, fb_sec, mf, n_stars_gold = calcula_probabilidades(data[data.dist<1.5], 'media',1)
```

```python
for cluster in df.index:
    massas =data.loc[cluster, 'mass']

    # Coeficientes angulares da IMF por partes
    a_h = df.loc[cluster, 'alpha_high_int']
    a_l = df.loc[cluster, 'alpha_low_int']
    Mc = df.loc[cluster, 'Mc_int']

    # Vamos fixar b1 = 1.0 arbitrariamente (o valor real não importa para médias ponderadas)
    b1 = 1.0
    b2 = b1 + (a_h - a_l) * Mc  # garantindo continuidade

    # Definindo a FMI por partes
    def fmi(x):
        return np.where(x <= Mc, a_l * x + b1, a_h * x + b2)

    # Calculando os pesos
    pesos = fmi(massas)


    pesos = np.where(pesos > 0, pesos, 1e-10)
    
    df.loc[cluster, 'mean_stellar_mass'] = np.sum(massas * pesos) / np.sum(pesos)

```
