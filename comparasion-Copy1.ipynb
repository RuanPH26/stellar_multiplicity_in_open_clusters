```python
def fb_donada(df):
    if isinstance(df, pd.Series):
        df = df.to_frame().T  # transforma em DataFrame de 1 linha
    sistemas_multiplos = len(df[df.q>0.6])
    sistemas_totais = len(df)+len(df[df.q<=0.6])
    return sistemas_multiplos/sistemas_totais

def fb_jedhav(df):
    if isinstance(df, pd.Series):
        df = df.to_frame().T  # transforma em DataFrame de 1 linha
    sistemas_multiplos = len(df[df.q>0.6])
    sistemas_totais = len(df)+len(df[df.q<=0.6])
    return sistemas_multiplos/sistemas_totais

def fb_jiang(df):
    if isinstance(df, pd.Series):
        df = df.to_frame().T  # transforma em DataFrame de 1 linha
    sistemas_multiplos = len(df[df.q>0.5])
    sistemas_totais = len(df)+len(df[df.q<=0.5])
    return sistemas_multiplos/sistemas_totais
```

```python
#MIMO
aglomerados = ['ASCC_21','Trumpler_3','Roslund_6','NGC_2287',\
              'NGC_2482',]
fb = [0.4,0.46,0.41,0.30,0.44]
e_fb = [0.03,0.05,0.03,0.03,0.06]

def fb_mimo(df):
    if isinstance(df, pd.Series):
        df = df.to_frame().T  # transforma em DataFrame de 1 linha
    sistemas_multiplos = len(df[df.q>0.2])
    sistemas_totais = len(df)+len(df[df.q<=0.2])
    return sistemas_multiplos/sistemas_totais

for i in range(0,len(aglomerados)):
    comparacao.loc[aglomerados[i], 'fb_mimo'] = fb_mimo(data.loc[aglomerados[i]])
    comparacao.loc[aglomerados[i], 'fb_mimo_ref'] = fb[i]
    comparacao.loc[aglomerados[i], 'er_fb_mimo_ref'] = e_fb[i]
    comparacao.loc[aglomerados[i], 'er_fb_mimo'] = comparacao.loc[aglomerados[i], 'fb_mimo']*np.sqrt(1/len(data.loc[aglomerados[i]])+1/len(data[data.q>0].loc[aglomerados[i]]))
```

```python
def corrigir_erro(fb, e_fb):
    if fb + e_fb > 1:
        return fb -1
    else:
        return e_fb
col = ['fb_donada', 'fb_jedhav', 'fb_jiang', 'fb_niu', 'fb_mimo']
for i in range(0, len(col)):
    comparacao['er_'+col[i]+'_s'] = comparacao.apply(lambda row: corrigir_erro(row[col[i]], row['er_'+col[i]]), axis=1)
```
