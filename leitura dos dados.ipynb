```python
import pandas as pd
import numpy as np
import os
import re
```

```python
df = pd.read_csv('Dados/log-results-eDR3-MF_integrada.csv', sep=';')
```

```python
list_arquivos = os.listdir('membership_data_edr3')
suffix = '_data_stars.npy'
clusters = [nome.removesuffix(suffix) for nome in list_arquivos]
commun_clusters = [cluster for cluster in clusters if cluster in set(df['Cluster'])]
```

```python
len(commun_clusters)
```

```python
data = pd.DataFrame()
for cluster in commun_clusters:
    df_cluster =  pd.DataFrame(np.load('membership_data_edr3/'+cluster+suffix))
    df_cluster.to_csv('Aglomerados/'+cluster+'.csv')
    df_cluster['Cluster'] = cluster
    data = pd.concat([data,df_cluster])
    
```

```python
for cluster in df['Cluster'].values:
    if cluster not in commun_clusters:
        print(cluster)
```
